```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fname = 'input_day10.txt'
```

```python
pos = []
vels = []
with open(fname) as f:
    while True:
        d = f.readline()
        if len(d)>0:
            dataline = [d[10:16], d[17:24], d[-8:-6], d[-4:-2]]
            dataline = [int(ele) for ele in dataline]
            pos.append(dataline[:2])
            vels.append(dataline[2:])
        else:
            break

pos_ar = np.array(pos)
vel_ar = np.array(vels)
```

```python
#find when they're closest together
xrange = []
pos_ar = np.array(pos)
for i in range(20000):
    xrange.append(np.abs(np.max(pos_ar[:,0]) - np.min(pos_ar[:,0])))
    pos_ar = pos_ar + vel_ar
```

```python
np.argmin(xrange)
```

```python
plt.plot(xrange)
```

```python
pos_ar = np.array(pos)

i = np.argmin(xrange)
    
pos_ar = np.array(pos) + i*vel_ar
plt.axis('equal')
plt.scatter(pos_ar[:,0], -pos_ar[:,1]) #message is upside down without the minus signs
plt.show()
```
